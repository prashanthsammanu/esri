```python
# List of columns to exclude from aggregation
exclude_columns = ['GEO_ID', 'state', 'tract', 'block group', 'county']

# Set 'county' as the index if it's not already
if 'county' not in combined_data.index.names:
    combined_data.set_index('county', inplace=True)

# Select only the columns to be aggregated
columns_to_aggregate = [col for col in combined_data.columns if col not in exclude_columns]

# Group the data by 'county' and aggregate only the specified columns
aggregated_data = combined_data[columns_to_aggregate].groupby('county').sum()

# Reset index to turn grouped 'county' back into a column
aggregated_data.reset_index(inplace=True)

# Rename the 'county' column to 'countyfp'
aggregated_data.rename(columns={'county': 'countyfp'}, inplace=True)

# Drop the 'NAME' column if it exists
if 'NAME' in aggregated_data.columns:
    aggregated_data.drop(columns=['NAME'], inplace=True)

# Save the aggregated data to a CSV file
aggregated_data.to_csv('aggregated_census_data.csv', index=False)
```

```python
# Read the CSV file into a DataFrame
csv_file = r"C:\Users\psammanu\Documents\esri\census_data.csv"
df = pd.read_csv(csv_file)

# Rename the "tract" column to "tracte"
df.rename(columns={"tract": "tracte"}, inplace=True)

# Add leading zeros to the "tracte" column to make it 6 digits long
df['tracte'] = df['tracte'].apply(lambda x: str(x).zfill(6))

# Group the DataFrame by the "tracte" column and aggregate values
aggregated_df = df.groupby("tracte").sum().reset_index()

# Remove the "NAME" column from the aggregated DataFrame
aggregated_df.drop(columns=["NAME"], inplace=True)

# Save the aggregated DataFrame to a new CSV file
aggregated_csv_file = "aggregated_censustracts.csv"
aggregated_df.to_csv(aggregated_csv_file, index=False)
```
